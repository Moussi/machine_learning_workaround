```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import SGDRegressor
```

# 1. Dataset 
dataset $(x, y)$ qui contient 100 exemples, et une seule variable $x$


```python
np.random.seed(0)
x, y = make_regression(n_samples=100, n_features=1, noise=10)
```

```python
print(f"x => {x.shape}, y => {y.shape}")
```

```python
plt.scatter(x,y)
```

```python
model = SGDRegressor(max_iter=1000, eta0=0.01)
model.fit(x, y)
```

```python
print("Determination Coef = ", model.score(x, y))
```

# Training

```python
predictions = model.predict(x)
plt.scatter(x, y)
plt.plot(x, predictions, c = 'r')
```

# Plynomial degree ==3

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)
x, y = make_regression(n_samples=100, n_features=2, noise=0.2)
y = y**2
```

```python
plt.scatter(x[:,1], y, c = y)
```

```python
poly = PolynomialFeatures(degree=3, include_bias=True)
poly_feature = poly.fit_transform(x.reshape(-2, 2))

model = LinearRegression()
model.fit(poly_feature, y)
```

```python
print("Determination Coef = ", model.score(poly_feature, y))
```

```python
plt.scatter(x[:,0], model.predict(poly_feature), c = 'r')
plt.scatter(x[:, 0], y)
```

```python
plt.scatter(x[:,1], model.predict(poly_feature), c = 'r')
plt.scatter(x[:, 1], y)
```

# Plynomial degree ==2

LinearRegression Uses normal method (not efficient for large datasets) Vs SGDRegressor uses gradient descend

```python
np.random.seed(0)
x, y = make_regression(n_samples=100, n_features=1, noise=10)
y = y**2
```

```python
plt.scatter(x, y)
```

```python
poly = PolynomialFeatures(degree=2, include_bias=False)
x = poly.fit_transform(x)
```

```python
model = SGDRegressor(max_iter=1000, eta0=0.01)
```

```python
model.fit(x, y)
```

```python
print("Determination Coef = ", model.score(x, y))
```

```python
plt.scatter(x[:, 0], y)
plt.scatter(x[:, 0], model.predict(x), c = 'r')
```
