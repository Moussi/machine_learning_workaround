```python
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
digits = load_digits()
```

```python
type(digits)
```

```python
X = digits.data
y = digits.target
```

```python
X.shape
```

```python
plt.imshow(digits['images'][0], cmap='gray')
```

```python
model = KNeighborsClassifier()
model.fit(X,y)
model.score(X, y)
```

```python
from PIL import Image
img = Image.open('four.png').resize((8, 8))

four = np.asarray(img)
plt.imshow(four)
```

```python
test = four[:,:,0].reshape(1, -1)
print(four[:,:,0].reshape(1, -1).shape)
model.predict(test)

```
