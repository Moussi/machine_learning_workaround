```python
import numpy as np
from sklearn.datasets import make_classification, load_iris
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
```

```python
np.random.seed(0)
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1)
```

```python
plt.figure(num=None, figsize=(8,6))
plt.xlabel('X0')
plt.ylabel('X1')
plt.scatter(X[:,0], X[:,1], c = y, marker='o', edgecolors='k')
```

```python
model = SGDClassifier(max_iter=1000, eta0=0.001, loss='log')
```

```python
model.fit(X, y)
```

```python
print('score:', model.score(X, y))
```

```python
h = .02
colors = "bry"
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
print(x_min, " ",x_max, " ", y_min, " ", y_max)
```

```python
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
plt.axis('tight')
for i, color in zip(model.classes_, colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=color, cmap=plt.cm.Paired, edgecolor='black', s
    =20)
```
